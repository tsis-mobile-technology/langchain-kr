````python
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 한국어 BERT 모델 로드하기 (예: klue/bert-base)
model_name = "klue/bert-base"  # 성능이 좋은 한국어 모델로 변경
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)  # 레이블 개수에 맞게 수정

# GPU 사용 가능 여부 확인
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)  # 모델을 GPU로 이동 (가능한 경우)

# 텍스트 분류하기
text = "이것은 예시 문장입니다."  # 한국어 텍스트 입력
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)  # 패딩 및 길이 제한 추가, GPU로 이동
outputs = model(**inputs)

# 예측 결과 출력하기
logits = outputs.logits
probabilities = F.softmax(logits, dim=-1)
predicted_class_id = logits.argmax(-1)

print("Logits:", logits)
print("Probabilities:", probabilities)
print("Prediction (Class ID):", predicted_class_id)
print(inputs)
````

## 코드 설명:

### 모델 실행:

output_hidden_states=True 옵션을 통해 각 레이어의 hidden state를 저장합니다.
output_attentions=True 옵션을 통해 각 레이어의 attention 값을 저장합니다.
각 레이어 출력값 확인:

outputs.hidden_states에는 각 레이어의 hidden state가 리스트 형태로 저장되어 있습니다.
outputs.attentions에는 각 레이어의 attention 값이 리스트 형태로 저장되어 있습니다.
특정 레이어 출력값 확인:

hidden_states[-1]은 마지막 레이어의 hidden state를 나타냅니다.
attentions[0][0]은 첫 번째 레이어의 첫 번째 attention head를 나타냅니다.
예측 결과 출력:

logits, probabilities, predicted_class_id를 출력하여 최종 예측 결과를 확인합니다.
주의 사항:

각 레이어의 출력값은 크기가 매우 크므로, 필요한 부분만 선택적으로 출력하는 것이 좋습니다.
어텐션 값은 각 토큰 간의 관계를 나타내는 행렬 형태로 저장되어 있으며, 시각화 도구를 사용하면 더 쉽게 이해할 수 있습니다.

````python
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델 로드 (한국어 모델 예시: klue/bert-base)
model_name = "klue/bert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 입력 텍스트
text = "이것은 예시 문장입니다."

# 토크나이징 및 모델 입력 준비
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

# 모델 실행 (각 레이어의 출력값 저장)
with torch.no_grad():
    outputs = model(**inputs, output_hidden_states=True, output_attentions=True)

# 각 레이어 출력값 확인
hidden_states = outputs.hidden_states
attentions = outputs.attentions

# 특정 레이어 출력값 확인 (예: 마지막 레이어)
last_hidden_state = hidden_states[-1]
print("Last Hidden State:", last_hidden_state)

# 어텐션 값 확인 (예: 첫 번째 레이어의 첫 번째 어텐션 헤드)
first_attention = attentions[0][0]
print("First Attention Head:", first_attention)

# 예측 결과 출력
logits = outputs.logits
probabilities = F.softmax(logits, dim=-1)
predicted_class_id = logits.argmax(-1)
print("Logits:", logits)
print("Probabilities:", probabilities)
print("Predicted Class ID:", predicted_class_id)
````

## 주요 변경 사항:

### Hidden State 출력:
tokenizer.convert_ids_to_tokens()를 사용하여 토큰 ID를 원래 문자열로 변환합니다.
각 토큰 문자열과 해당 벡터 값을 함께 출력합니다.
Attention 출력:
각 토큰에 대한 어텐션 값을 행렬 형태로 출력합니다.
이제 수정된 코드는 hidden state와 attention 값을 사람이 이해하기 쉬운 문자열 형태로 출력합니다. 각 토큰의 hidden state 벡터와 어텐션 값을 분석하여 모델이 입력 텍스트를 어떻게 처리하고 분류했는지 더 자세히 파악할 수 있습니다.

````python
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델 로드 (한국어 모델 예시: klue/bert-base)
model_name = "klue/bert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 입력 텍스트
text = "이것은 예시 문장입니다."

# 토크나이징 및 모델 입력 준비
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

# 모델 실행 (각 레이어의 출력값 저장)
with torch.no_grad():
    outputs = model(**inputs, output_hidden_states=True, output_attentions=True)

# 각 레이어 출력값 확인
hidden_states = outputs.hidden_states
attentions = outputs.attentions

# 특정 레이어 출력값 확인 (예: 마지막 레이어)
last_hidden_state = hidden_states[-1]
print("Last Hidden State:")
for token_index, token_vector in enumerate(last_hidden_state[0]):  # 첫 번째 문장에 대한 벡터 출력
    token_str = tokenizer.convert_ids_to_tokens(inputs.input_ids[0][token_index])  # 토큰 ID를 문자열로 변환
    print(f"{token_str}: {token_vector}")  # 토큰 문자열과 벡터 값 출력

# 어텐션 값 확인 (예: 첫 번째 레이어의 첫 번째 어텐션 헤드)
first_attention = attentions[0][0]
print("\nFirst Attention Head:")
for i, row in enumerate(first_attention[0]):  # 첫 번째 문장에 대한 어텐션 행렬 출력
    print(f"Token {i+1}: {row}")

# 예측 결과 출력
logits = outputs.logits
probabilities = F.softmax(logits, dim=-1)
predicted_class_id = logits.argmax(-1)

print("\nLogits:", logits)
print("Probabilities:", probabilities)
print("Predicted Class ID:", predicted_class_id)
````

### 핵심 변경 사항:

.item() 추가: inputs.input_ids[0][token_index].item()을 사용하여 0차원 텐서를 정수 값으로 변환합니다. tokenizer.convert_ids_to_tokens() 함수는 정수 또는 정수 리스트를 입력으로 받기 때문입니다.
이렇게 수정하면 각 토큰의 ID를 정수로 변환하여 tokenizer.convert_ids_to_tokens() 함수에 전달하므로, "iteration over a 0-d tensor" 오류가 발생하지 않습니다.

````python
# 네, 문제는 tokenizer.convert_ids_to_tokens() 함수에 0차원 텐서를 전달하는 데 있습니다. 이 함수는 1차원 텐서(리스트)를 입력으로 받아야 합니다.
# 다음과 같이 수정하면 문제를 해결할 수 있습니다.
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델 로드 (한국어 모델 예시: klue/bert-base)
model_name = "klue/bert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# 입력 텍스트
text = "이것은 예시 문장입니다."

# 토크나이징 및 모델 입력 준비
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

# 모델 실행 (각 레이어의 출력값 저장)
with torch.no_grad():
    outputs = model(**inputs, output_hidden_states=True, output_attentions=True)

# 각 레이어 출력값 확인
hidden_states = outputs.hidden_states
attentions = outputs.attentions

# 특정 레이어 출력값 확인 (예: 마지막 레이어)
last_hidden_state = hidden_states[-1]
print("Last Hidden State:")
for token_index, token_vector in enumerate(last_hidden_state[0]):  # 첫 번째 문장에 대한 벡터 출력
    token_id = inputs.input_ids[0][token_index].item()  # 텐서를 정수 값으로 변환
    token_str = tokenizer.convert_ids_to_tokens(token_id)  # 정수 값을 문자열로 변환
    print(f"{token_str}: {token_vector}") 

# 어텐션 값 확인 (예: 첫 번째 레이어의 첫 번째 어텐션 헤드)
first_attention = attentions[0][0]
print("\nFirst Attention Head:")
for i, row in enumerate(first_attention[0]):
    print(f"Token {i+1}: {row}")

# 예측 결과 출력
logits = outputs.logits
probabilities = F.softmax(logits, dim=-1)
predicted_class_id = logits.argmax(-1)

print("\nLogits:", logits)
print("Probabilities:", probabilities)
print("Predicted Class ID:", predicted_class_id)
````
